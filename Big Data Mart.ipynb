```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('../DataSets/B_train.csv')
df
```

```python
df.isnull().sum()
```

```python
df.Outlet_Size.value_counts()
```

```python
df.Item_Weight=df.Item_Weight.fillna(12.8)
df.Outlet_Size=df.Outlet_Size.fillna('Unknown')
```

```python
df.dtypes
```

```python
dt=df.drop(['Item_Identifier','Outlet_Identifier','Outlet_Establishment_Year'],axis=1)
```

```python
dt
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
dt.Item_Type=label.fit_transform(dt.Item_Type)
dt.Outlet_Size=label.fit_transform(dt.Outlet_Size)
dt.Outlet_Location_Type=label.fit_transform(dt.Outlet_Location_Type)
dt.Outlet_Type=label.fit_transform(dt.Outlet_Type)
dt.Item_Fat_Content=label.fit_transform(dt.Item_Fat_Content)
```

```python
dt
```

```python
x=dt.drop('Item_Outlet_Sales',axis=1).values
x
```

```python
y=dt['Item_Outlet_Sales'].values
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2=r2_score(ypred,ytest)
r2
```

```python
mse=mean_squared_error(ypred,ytest)
mse
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```
